```python
# Using list comprehension ,list related functions and otherwise, create following list: [[1, 2, 3, 4, 5, 6, 7], [1, 4, 9, 16, 25, 36, 49], [1, 8, 27, 64, 125, 216,343]]

def list_comprehension(n):

    numbers = [n for n in range(1, n)]
    squares = [n**2 for n in range(1, n)]
    cubes = [n**3 for n in range(1, n)]
   
    return [numbers, squares, cubes]
     
n = 8
result = list_comprehension(n)
print(result)
```

```python

# Write a function which takes input 3 dictionaries and returns a new dictionary which has all key value pairs from those three dictionaries
 

# dic1={1:10, 2:20}
# dic2={3:30, 4:40}
# dic3={5:50, 6:60}
# dic4 = {}
# for d in (dic1, dic2, dic3): 
#     dic4.update(d)
# print(dic4)

# inserted of combine of tuple pass it as a key **kw

def combined_dic(combined):
    dic4 = {}
    for d in (combined): 
        dic4.update(d)
    return dic4
    
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

combined = (dic1, dic2, dic3)
dic = combined_dic(combined)    
    
print(dic)
```

```python
# Write a function which takes input n and produces first n elements of a fibonacci series


# def fibonacci_serie(n):
#     if (n <= 1):
#         return n

#     return fibonacci_serie(n-1) + fibonacci_serie(n-2)


# n = 8
# result = fibonacci_serie(n)

# print(result)


def fibonacci(n):
    if (n <= 1):
        return n

    return fibonacci(n-1) + fibonacci(n-2)

def get_fibonacci_ser(n):
    fibonacci_series = []
    for item in range(n):
        num = fibonacci(item)
        fibonacci_series.append(num)

    return fibonacci_series

n = 9
result = get_fibonacci_ser(n)


print(result)


# approch 2: 
# print series 
# input   8
# output 0 1 1 2 3 5 8 13 21 

# def fibonacci_rev(n):
#     if n == 1:
#         return [0]
#     elif n == 2:
#         return [0,1]
#     else:
#         x = fibonacci_rev(n-1)
#         x.append(sum(x[:-3:-1]))
#         return x

# x = fibonacci_rev(8)

# print(x)
```
